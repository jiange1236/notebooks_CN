````python
from unsloth import FastLanguageModel  # llmsçš„FastVisionModel
import torch
max_seq_length = 2048  # é€‰æ‹©ä»»ä½•ï¼æˆ‘ä»¬åœ¨å†…éƒ¨è‡ªåŠ¨æ”¯æŒç»³ç´¢æ‰©å±•ï¼
load_in_4bit = True  # ä½¿ç”¨4ä½é‡åŒ–æ¥å‡å°‘å†…å­˜ä½¿ç”¨é‡ã€‚å¯ä»¥æ˜¯é”™è¯¯çš„ã€‚

# 4ä½é¢„é‡åŒ–æ¨¡å‹æˆ‘ä»¬æ”¯æŒ4å€ä¸‹è½½ + no oomsã€‚
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",  # Llama-3.1 2å€
    "unsloth/Mistral-Small-Instruct-2409",  # MISTRAL 22B 2Xå¿«ï¼
    "unsloth/Phi-4",  # PHI-4 2Xå¿«ï¼
    "unsloth/Phi-4-unsloth-bnb-4bit",  # phi-4ä¸æ•´é½çš„åŠ¨æ€4ä½å®šé‡
    "unsloth/gemma-2-9b-bnb-4bit",  # Gemma 2Xå¿«ï¼
    "unsloth/Qwen2.5-7B-Instruct-bnb-4bit"  # QWEN 2.5 2å€å¿«ï¼
    "unsloth/Llama-3.2-1B-bnb-4bit",  # æ–°çš„ï¼ Llama 3.2å‹å·
    "unsloth/Llama-3.2-1B-Instruct-bnb-4bit",
    "unsloth/Llama-3.2-3B-bnb-4bit",
    "unsloth/Llama-3.2-3B-Instruct-bnb-4bit",
]  # https://docs.unsloth.ai/get-started/all-our-modelsçš„æ›´å¤šå‹å·

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-4",
    max_seq_length = max_seq_length,
    load_in_4bit = load_in_4bit,
    # token =â€œ hf _...â€ï¼Œï¼ƒå¦‚æœä½¿ç”¨å…ƒæ¨¡å‹ï¼Œä¾‹å¦‚meta-llama/llama-2-7b-hf
)
````

ç°åœ¨ï¼Œæˆ‘ä»¬ä¸ºå‚æ•°æœ‰æ•ˆçš„èŠ¬ç‰¹æ·»åŠ LORAé€‚é…å™¨ - è¿™ä½¿æˆ‘ä»¬åªèƒ½æœ‰æ•ˆåœ°è®­ç»ƒæ‰€æœ‰å‚æ•°çš„1ï¼…ã€‚

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # é€‰æ‹©ä»»ä½•æ•°å­—> 0ï¼å»ºè®®8ã€16ã€32ã€64ã€128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # æ”¯æŒä»»ä½•äººï¼Œä½†= 0å·²ä¼˜åŒ–
    bias = "none",    # æ”¯æŒä»»ä½•äººï¼Œä½†æ˜¯=â€œæ— â€
    # [NEW]â€œ Unsplothâ€ä½¿ç”¨çš„VRAMå‡å°‘äº†30ï¼…ï¼Œé€‚åˆ2å€æ‰¹é‡çš„å°ºå¯¸ï¼
    use_gradient_checkpointing = "unsloth", # åœ¨å¾ˆé•¿çš„èƒŒæ™¯ä¸‹çœŸå®æˆ–â€œä¸æ•´é½â€
    random_state = 3407,
    use_rslora = False,  # æˆ‘ä»¬æ”¯æŒæ’åç¨³å®šçš„æ´›æ‹‰
    loftq_config = None, # å’ŒLoftq
)
````

###æ•°æ®å‡†å¤‡
ç°åœ¨ï¼Œæˆ‘ä»¬å°†â€œ Phi-4â€æ ¼å¼ç”¨äºå¯¹è¯é£æ ¼çš„ç™»å½•ã€‚æˆ‘ä»¬åœ¨ShareGptæ ·å¼ä¸­ä½¿ç”¨XX_Markdown_link_xxæ•°æ®é›†ã€‚ä½†æ˜¯æˆ‘ä»¬å°†å…¶è½¬æ¢ä¸ºHuggingfaceçš„æ­£å¸¸å¤šå…ƒæ ¼å¼`ï¼ˆâ€œè§’è‰²â€ï¼Œâ€œ contentâ€ï¼‰`è€Œä¸æ˜¯`ï¼ˆâ€œæ¥è‡ªâ€ï¼Œâ€œ valueâ€ï¼‰`/ phi-4å‘ˆç°å¤šè½¬å‘å¯¹è¯å¦‚ä¸‹ï¼š

```
<|im_start|>user<|im_sep|>Hello!<|im_end|>
<|im_start|>assistant<|im_sep|>Hi! How can I help?<|im_end|>
<|im_start|>user<|im_sep|>What is 2+2?<|im_end|>
```

æˆ‘ä»¬ä½¿ç”¨â€œ get_chat_templateâ€å‡½æ•°è·å–æ­£ç¡®çš„èŠå¤©æ¨¡æ¿ã€‚æˆ‘ä»¬æ”¯æŒ`Zephyrï¼ŒChatmlï¼ŒMistralï¼ŒLlamaï¼Œç¾Šé©¼ï¼ŒVicunaï¼ŒVicuna_oldï¼ŒPhi3ï¼ŒPhi4ï¼ŒLllame3`ç­‰ã€‚

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "phi-4",
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [
        tokenizer.apply_chat_template(
            convo, tokenize = False, add_generation_prompt = False
        )
        for convo in convos
    ]
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("mlabonne/FineTome-100k", split = "train")
````

ç°åœ¨ï¼Œæˆ‘ä»¬ä½¿ç”¨â€œ standardize_sharegptâ€å°†ShareGptæ ·å¼æ•°æ®é›†è½¬æ¢ä¸ºHuggingFaceçš„é€šç”¨æ ¼å¼ã€‚è¿™å°†æ•°æ®é›†ä»çœ‹èµ·æ¥åƒï¼š
```
{"from": "system", "value": "You are an assistant"}
{"from": "human", "value": "What is 2+2?"}
{"from": "gpt", "value": "It's 4."}
```
åˆ°
```
{"role": "system", "content": "You are an assistant"}
{"role": "user", "content": "What is 2+2?"}
{"role": "assistant", "content": "It's 4."}
```

````python
from unsloth.chat_templates import standardize_sharegpt

dataset = standardize_sharegpt(dataset)
dataset = dataset.map(
    formatting_prompts_func,
    batched=True,
)
````

<a name =â€œ trainâ€> </a>
###è®­ç»ƒæ¨¡å‹
ç°åœ¨ï¼Œè®©æˆ‘ä»¬ä½¿ç”¨huggingface trlçš„â€œ sfttrainerâ€ï¼è¿™é‡Œæ›´å¤šæ–‡æ¡£ï¼š[TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer)ã€‚æˆ‘ä»¬æ‰§è¡Œ60ä¸ªæ­¥éª¤æ¥åŠ å¿«é€Ÿåº¦ï¼Œä½†æ˜¯æ‚¨å¯ä»¥å°†`num_train_epochs = 1`è®¾ç½®ä¸ºå®Œæ•´è¿è¡Œï¼Œç„¶åå…³é—­`max_steps = noneâ€œã€‚æˆ‘ä»¬è¿˜æ”¯æŒTRLçš„â€œ Dpotrainerâ€ï¼

````python
from trl import SFTConfig, SFTTrainer
from transformers import DataCollatorForSeq2Seq
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer),
    packing = False, # å¯ä»¥ä½¿çŸ­åºåˆ—æ›´å¿«åœ°è®­ç»ƒ5å€ã€‚
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1ï¼Œï¼ƒå°†å…¶è®¾ç½®ä¸º1ä¸ªå®Œæ•´çš„è®­ç»ƒè¿è¡Œã€‚
        max_steps = 30,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # å°†å…¶ç”¨äºWandbç­‰
    ),
)
````

æˆ‘ä»¬è¿˜ä½¿ç”¨Unsplothçš„â€œ Train_on_completionsâ€æ–¹æ³•ä»…è®­ç»ƒåŠ©æ‰‹è¾“å‡ºï¼Œè€Œå¿½ç•¥ç”¨æˆ·è¾“å…¥çš„æŸå¤±ã€‚

````python
from unsloth.chat_templates import train_on_responses_only

trainer = train_on_responses_only(
    trainer,
    instruction_part="<|im_start|>user<|im_sep|>",
    response_part="<|im_start|>assistant<|im_sep|>",
)
````
