**Êàë‰ª¨Ê≠£Âú®‰ΩøÁî®OpenAiÁöÑMXFP4 TritonÂÜÖÊ†∏‰∏é‰∏çË£§ÁöÑÂÜÖÊ†∏Áõ∏ÁªìÂêàÔºÅ**

```python
from unsloth import FastLanguageModel
import torch

# 4‰ΩçÈ¢ÑÈáèÂåñÊ®°ÂûãÊàë‰ª¨ÊîØÊåÅ4ÂÄç‰∏ãËΩΩ + no ooms„ÄÇ
fourbit_models = [
    "unsloth/gpt-oss-20b-unsloth-bnb-4bit", # ‰ΩøÁî®BITSANDBYTES 4BITÈáèÂåñÁöÑ20BÊ®°Âûã
    "unsloth/gpt-oss-120b-unsloth-bnb-4bit",
    "unsloth/gpt-oss-20b", # ‰ΩøÁî®MXFP4Ê†ºÂºèÁöÑ20BÊ®°Âûã
    "unsloth/gpt-oss-120b",
] # https://huggingface.co/unslothÁöÑÊõ¥Â§öÂûãÂè∑

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gpt-oss-20b",
    dtype = None, # Ê≤°ÊúâËá™Âä®Ê£ÄÊµã
    max_seq_length = 4096, # ‰∏∫ÈïøÁØáÂ∞èËØ¥ÈÄâÊã©‰ªª‰ΩïÔºÅ
    load_in_4bit = False,  # 4‰ΩçÈáèÂåñ‰ª•ÂáèÂ∞ëÂÜÖÂ≠ò
    full_finetuning = False, # [newÔºÅ]Êàë‰ª¨Áé∞Âú®ÊúâÂÆåÊï¥ÁöÑÂ°´ÂÖÖÔºÅ
    # token =‚Äú hf _...‚ÄùÔºåÔºÉÂ¶ÇÊûú‰ΩøÁî®Èó®ÊéßÊ®°ÂûãÔºåËØ∑‰ΩøÁî®‰∏Ä‰∏™
)
```

### Êé®ÁêÜÂä™Âäõ
OpenAIÁöÑ‚Äú GPT-oss‚ÄùÊ®°ÂûãÂåÖÂê´‰∏ÄÈ°πÂäüËÉΩÔºåËØ•ÂäüËÉΩÂÖÅËÆ∏Áî®Êà∑Ë∞ÉÊï¥Ê®°ÂûãÁöÑ‚ÄúÊé®ÁêÜÂ∑•‰Ωú‚Äù„ÄÇËøô‰ΩøÊÇ®ÂèØ‰ª•ÊéßÂà∂Ê®°ÂûãÁöÑÊÄßËÉΩ‰∏éÂÖ∂ÂìçÂ∫îÈÄüÂ∫¶ÔºàÂª∂ËøüÔºâ‰πãÈó¥ÁöÑÊùÉË°°ÔºåËØ•ÈÄüÂ∫¶ÔºàÂª∂ËøüÔºâÈÄöËøáËØ•Ê®°ÂûãÂ∞Ü‰ΩøÁî®ÁöÑ‰ª§ÁâåÊï∞ÈáèÊù•ÊÄùËÄÉ„ÄÇ

‚Äú GPT-oss‚ÄùÊ®°ÂûãÊèê‰æõ‰∫Ü‰∏â‰∏™‰∏çÂêåÁ∫ßÂà´ÁöÑÊé®ÁêÜÂ∑•‰ΩúÔºåÊÇ®ÂèØ‰ª•ÈÄâÊã©Ôºö

***‰Ωé**ÔºöÈíàÂØπÈúÄË¶ÅÈùûÂ∏∏Âø´ÈÄüÂìçÂ∫îÂπ∂‰∏î‰∏çÈúÄË¶ÅÂ§çÊùÇÁöÑÂ§öÊ≠•Êé®ÁêÜÁöÑ‰ªªÂä°ËøõË°å‰∫Ü‰ºòÂåñ„ÄÇ
***‰∏≠**ÔºöÊÄßËÉΩ‰∏éÈÄüÂ∫¶‰πãÈó¥ÁöÑÂπ≥Ë°°„ÄÇ
***È´ò**Ôºö‰∏∫ÈúÄË¶ÅÂÆÉÁöÑ‰ªªÂä°Êèê‰æõÊúÄÂº∫ÁöÑÊé®ÁêÜÊÄßËÉΩÔºåÂ∞ΩÁÆ°Ëøô‰ºöÂØºËá¥Êõ¥È´òÁöÑÂª∂Ëøü„ÄÇ

```python
from transformers import TextStreamer

messages = [
    {"role": "user", "content": "Solve x^5 + 3x^4 - 10 = 3."},
]
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True,
    return_tensors = "pt",
    return_dict = True,
    reasoning_effort = "low", # **Êñ∞ÔºÅ**Â∞ÜÊé®ÁêÜÂä™ÂäõËÆæÁΩÆ‰∏∫‰ΩéÔºå‰∏≠ÊàñÈ´ò
).to(model.device)

_ = model.generate(**inputs, max_new_tokens = 512, streamer = TextStreamer(tokenizer))
```
