````python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # é€‰æ‹©ä»»ä½•ï¼æˆ‘ä»¬åœ¨å†…éƒ¨è‡ªåŠ¨æ”¯æŒç»³ç´¢æ‰©å±•ï¼
dtype = None # æ²¡æœ‰è‡ªåŠ¨æ£€æµ‹ã€‚ Tesla T4ï¼ŒV100ï¼ŒBfloat16çš„float16 for Ampere+
load_in_4bit = True # ä½¿ç”¨4ä½é‡åŒ–æ¥å‡å°‘å†…å­˜ä½¿ç”¨é‡ã€‚å¯ä»¥æ˜¯é”™è¯¯çš„ã€‚

# 4ä½é¢„é‡åŒ–æ¨¡å‹æˆ‘ä»¬æ”¯æŒ4å€ä¸‹è½½ + no oomsã€‚
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 15ä¸‡äº¿ä»£å¸å‹2å€ï¼
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # æˆ‘ä»¬è¿˜ä¸Šä¼ äº†405Bçš„4ä½ï¼
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # æ–°çš„Mistral 12b 2xå¿«ï¼
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral V3 2Xå¿«é€Ÿï¼
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # PHI-3.5 2xå¿«ï¼
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemma 2Xå¿«ï¼
] # https://huggingface.co/unslothçš„æ›´å¤šå‹å·

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Phi-3.5-mini-instruct",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token =â€œ hf _...â€ï¼Œï¼ƒå¦‚æœä½¿ç”¨å…ƒæ¨¡å‹ï¼Œä¾‹å¦‚meta-llama/llama-2-7b-hf
)
````

ç°åœ¨ï¼Œæˆ‘ä»¬æ·»åŠ Loraé€‚é…å™¨ï¼Œå› æ­¤æˆ‘ä»¬åªéœ€è¦æ›´æ–°æ‰€æœ‰å‚æ•°çš„1ï¼…è‡³10ï¼…ï¼

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # é€‰æ‹©ä»»ä½•æ•°å­—> 0ï¼å»ºè®®8ã€16ã€32ã€64ã€128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # æ”¯æŒä»»ä½•äººï¼Œä½†= 0å·²ä¼˜åŒ–
    bias = "none",    # æ”¯æŒä»»ä½•äººï¼Œä½†æ˜¯=â€œæ— â€
    # [NEW]â€œ Unsplothâ€ä½¿ç”¨çš„VRAMå‡å°‘äº†30ï¼…ï¼Œé€‚åˆ2å€æ‰¹é‡çš„å°ºå¯¸ï¼
    use_gradient_checkpointing = "unsloth", # åœ¨å¾ˆé•¿çš„èƒŒæ™¯ä¸‹çœŸå®æˆ–â€œä¸æ•´é½â€
    random_state = 3407,
    use_rslora = False,  # æˆ‘ä»¬æ”¯æŒæ’åç¨³å®šçš„æ´›æ‹‰
    loftq_config = None, # å’ŒLoftq
)
````

æˆ‘ä»¬ä½¿ç”¨â€œ get_chat_templateâ€å‡½æ•°è·å–æ­£ç¡®çš„èŠå¤©æ¨¡æ¿ã€‚æˆ‘ä»¬æ”¯æŒâ€œ Zephyrï¼ŒChatmlï¼ŒMistralï¼ŒLlamaï¼Œç¾Šé©¼ï¼ŒVicunaï¼ŒVicuna_oldâ€å’Œæˆ‘ä»¬è‡ªå·±çš„ä¼˜åŒ–`'Unsploth'æ¨¡æ¿ã€‚

Note ShareGptä½¿ç”¨`{â€œ fromâ€ï¼šâ€œ humanâ€ï¼Œâ€œ valueâ€ï¼šâ€œ hiâ€}`è€Œä¸æ˜¯`{â€œ coleâ€ï¼šâ€œ userâ€ï¼šâ€œ userâ€ï¼Œâ€œ contentâ€ï¼šâ€œ hiâ€}`ï¼Œå› æ­¤æˆ‘ä»¬ä½¿ç”¨`æ˜ å°„'å°†å…¶æ˜ å°„ã€‚

å¯¹äºè¯¸å¦‚æ–°é¢–å†™ä½œä¹‹ç±»çš„æ–‡æœ¬å®Œæˆï¼Œè¯·å°è¯•æ­¤[here](https://github.com/unslothai/unsloth/wiki#train-on-completions--responses-only-do-not-train-on-inputs)-text_completion.ipynbï¼‰ã€‚

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "phi-3", # æ”¯æŒZephyrï¼ŒChatmlï¼ŒMistralï¼ŒLlamaï¼Œç¾Šé©¼ï¼ŒVicunaï¼ŒVicuna_oldï¼ŒUnsploth
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPTé£æ ¼
)

def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }
pass

from datasets import load_dataset
dataset = load_dataset("philschmid/guanaco-sharegpt-style", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
````

XX_Markdown_link_xxä¸Šçš„èŠå¤©æ¨¡æ¿ä¸Šçš„æ›´å¤šä¿¡æ¯

````python
unsloth_template = \
    "{{ bos_token }}"\
    "{{ 'You are a helpful assistant to the user\n' }}"\
    "{% for message in messages %}"\
        "{% if message['role'] == 'user' %}"\
            "{{ '>>> User: ' + message['content'] + '\n' }}"\
        "{% elif message['role'] == 'assistant' %}"\
            "{{ '>>> Assistant: ' + message['content'] + eos_token + '\n' }}"\
        "{% endif %}"\
    "{% endfor %}"\
    "{% if add_generation_prompt %}"\
        "{{ '>>> Assistant: ' }}"\
    "{% endif %}"
unsloth_eos_token = "eos_token"

if False:
    tokenizer = get_chat_template(
        tokenizer,
        chat_template = (unsloth_template, unsloth_eos_token,), # æ‚¨å¿…é¡»æä¾›æ¨¡æ¿å’ŒEOSä»¤ç‰Œ
        mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPTé£æ ¼
        map_eos_token = True, # åœ°å›¾<| im_end |> to </s>
    )
````

<a name =â€œ trainâ€> </a>
###è®­ç»ƒæ¨¡å‹
ç°åœ¨ï¼Œè®©æˆ‘ä»¬ä½¿ç”¨huggingface trlçš„â€œ sfttrainerâ€ï¼è¿™é‡Œæ›´å¤šæ–‡æ¡£ï¼š[TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer)ã€‚æˆ‘ä»¬æ‰§è¡Œ60ä¸ªæ­¥éª¤æ¥åŠ å¿«é€Ÿåº¦ï¼Œä½†æ˜¯æ‚¨å¯ä»¥å°†`num_train_epochs = 1`è®¾ç½®ä¸ºå®Œæ•´è¿è¡Œï¼Œç„¶åå…³é—­`max_steps = noneâ€œã€‚æˆ‘ä»¬è¿˜æ”¯æŒTRLçš„â€œ Dpotrainerâ€ï¼

````python
from trl import SFTConfig, SFTTrainer
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    packing = False, # å¯ä»¥ä½¿çŸ­åºåˆ—æ›´å¿«åœ°è®­ç»ƒ5å€ã€‚
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # å°†å…¶ç”¨äºWandbç­‰
    ),
)
````

````python
# @Titleæ˜¾ç¤ºå½“å‰å†…å­˜ç»Ÿè®¡
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
````

````python
trainer_stats = trainer.train()
````

````python
# @Titleæ˜¾ç¤ºæœ€ç»ˆå†…å­˜å’Œæ—¶é—´ç»Ÿè®¡
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
````

è®©æˆ‘ä»¬è¿è¡Œæ¨¡å‹ï¼ç”±äºæˆ‘ä»¬æ­£åœ¨ä½¿ç”¨`phi-3``ï¼Œè¯·ä½¿ç”¨``apply_chat_template''å°†add_generation_prompt`è®¾ç½®ä¸ºtrue`è¿›è¡Œæ¨ç†ã€‚

````python
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "phi-3", # æ”¯æŒZephyrï¼ŒChatmlï¼ŒMistralï¼ŒLlamaï¼Œç¾Šé©¼ï¼ŒVicunaï¼ŒVicuna_oldï¼ŒUnsploth
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPTé£æ ¼
)

FastLanguageModel.for_inference(model) # å¯ç”¨æœ¬åœ°2å€æ›´å¿«çš„æ¨ç†

messages = [
    {"from": "human", "value": "Continue the fibonnaci sequence: 1, 1, 2, 3, 5, 8,"},
]
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True, # å¿…é¡»æ·»åŠ ä¸€ä»£
    return_tensors = "pt",
).to("cuda")

outputs = model.generate(input_ids = inputs, max_new_tokens = 64, use_cache = True)
tokenizer.batch_decode(outputs)
````
