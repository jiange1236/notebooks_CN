````python
from synthetic_data_kit.core.save_as import convert_format
import os
import json

# Áõ¥Êé•ËÆæÁΩÆÊâÄÊúâÂèÇÊï∞
input_file = "data/cleaned/ai_meta_com_qa_pairs_cleaned.json"
format_type = "ft"  # OpenaiÂæÆË∞ÉÊ†ºÂºè
storage_format = "json"  # ÈªòËÆ§Â≠òÂÇ®Ê†ºÂºè

# ËÆæÁΩÆËæìÂá∫Ë∑ØÂæÑ
final_dir = "data/final"
# OS.MakedirsÔºàfinal_dirÔºåequent_ok = trueÔºâ
base_name = os.path.splitext(os.path.basename(input_file))[0]

# Á°ÆÂÆöËæìÂá∫Êñá‰ª∂Ë∑ØÂæÑ
if storage_format == "hf":
    output_path = os.path.join(final_dir, f"{base_name}_{format_type}_hf")
else:
    if format_type == "jsonl":
        output_path = os.path.join(final_dir, f"{base_name}.jsonl")
    else:
        output_path = os.path.join(final_dir, f"{base_name}_{format_type}.json")

# Âä†ËΩΩÈÖçÁΩÆÂ¶ÇÊûúÂèØÁî®
config = ctx.config if 'ctx' in locals() else None

try:
    # Áõ¥Êé•Ëá¥Áîµconvert_format
    result_path = convert_format(
        input_file,
        output_path,
        format_type,
        config,
        storage_format=storage_format
    )

    print(f"Converted to {format_type} format and saved to {result_path}")

    # ÊòæÁ§∫ËΩ¨Êç¢Êñá‰ª∂ÁöÑÂÜÖÂÆπ
    try:
        if os.path.isfile(result_path):
            with open(result_path, 'r') as f:
                output_content = f.read()
            print("\ nconvertedÂÜÖÂÆπÔºàÂâç500‰∏™Â≠óÁ¨¶ÔºâÔºö")
            print(output_content[:500] + "..." if len(output_content) > 500 else output_content)
        else:
            # ÂØπ‰∫éHFÊï∞ÊçÆÈõÜÔºåËøôÊòØ‰∏Ä‰∏™ÁõÆÂΩï
            print(f"\nSaved as HF dataset directory at {result_path}")
            if os.path.exists(os.path.join(result_path, "dataset_info.json")):
                with open(os.path.join(result_path, "dataset_info.json"), 'r') as f:
                    info = json.load(f)
                print(f"Dataset info: {info}")
    except Exception as e:
        print(f"Could not read converted file: {e}")

except Exception as e:
    print(f"Error: {e}")
````

````python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # ÈÄâÊã©‰ªª‰ΩïÔºÅÊàë‰ª¨Âú®ÂÜÖÈÉ®Ëá™Âä®ÊîØÊåÅÁª≥Á¥¢Êâ©Â±ïÔºÅ
dtype = None # Ê≤°ÊúâËá™Âä®Ê£ÄÊµã„ÄÇ Tesla T4ÔºåV100ÔºåBfloat16ÁöÑfloat16 for Ampere+
load_in_4bit = True # ‰ΩøÁî®4‰ΩçÈáèÂåñÊù•ÂáèÂ∞ëÂÜÖÂ≠ò‰ΩøÁî®Èáè„ÄÇÂèØ‰ª•ÊòØÈîôËØØÁöÑ„ÄÇ

# 4‰ΩçÈ¢ÑÈáèÂåñÊ®°ÂûãÊàë‰ª¨ÊîØÊåÅ4ÂÄç‰∏ãËΩΩ + no ooms„ÄÇ
fourbit_models = [
    "unsloth/mistral-7b-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.2-bnb-4bit",
    "unsloth/llama-2-7b-bnb-4bit",
    "unsloth/llama-2-13b-bnb-4bit",
    "unsloth/codellama-34b-bnb-4bit",
    "unsloth/tinyllama-bnb-4bit",
    "unsloth/gemma-7b-bnb-4bit", # Êñ∞ÁöÑGoogle 6‰∏á‰∫ø‰ª£Â∏Å2.5ÂÄçÔºÅ
    "unsloth/gemma-2b-bnb-4bit",
] # https://huggingface.co/unslothÁöÑÊõ¥Â§öÂûãÂè∑

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Llama-3.2-3B-Instruct", # ÈÄâÊã©‰ªª‰ΩïÔºÅ‰æãÂ¶ÇTeknium/OpenHermes-2.5-Mistral-7b
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token =‚Äú hf _...‚ÄùÔºåÔºÉÂ¶ÇÊûú‰ΩøÁî®ÂÖÉÊ®°ÂûãÔºå‰æãÂ¶Çmeta-llama/llama-2-7b-hf
)
````

Áé∞Âú®ÔºåÊàë‰ª¨Ê∑ªÂä†LoraÈÄÇÈÖçÂô®ÔºåÂõ†Ê≠§Êàë‰ª¨Âè™ÈúÄË¶ÅÊõ¥Êñ∞ÊâÄÊúâÂèÇÊï∞ÁöÑ1ÔºÖËá≥10ÔºÖÔºÅ

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # ÈÄâÊã©‰ªª‰ΩïÊï∞Â≠ó> 0ÔºÅÂª∫ËÆÆ8„ÄÅ16„ÄÅ32„ÄÅ64„ÄÅ128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # ÊîØÊåÅ‰ªª‰Ωï‰∫∫Ôºå‰ΩÜ= 0Â∑≤‰ºòÂåñ
    bias = "none",    # ÊîØÊåÅ‰ªª‰Ωï‰∫∫Ôºå‰ΩÜÊòØ=‚ÄúÊó†‚Äù
    # [NEW]‚Äú Unsploth‚Äù‰ΩøÁî®ÁöÑVRAMÂáèÂ∞ë‰∫Ü30ÔºÖÔºåÈÄÇÂêà2ÂÄçÊâπÈáèÁöÑÂ∞∫ÂØ∏ÔºÅ
    use_gradient_checkpointing = "unsloth", # Âú®ÂæàÈïøÁöÑËÉåÊôØ‰∏ãÁúüÂÆûÊàñ‚Äú‰∏çÊï¥ÈΩê‚Äù
    random_state = 3407,
    use_rslora = False,  # Êàë‰ª¨ÊîØÊåÅÊéíÂêçÁ®≥ÂÆöÁöÑÊ¥õÊãâ
    loftq_config = None, # ÂíåLoftq
)
````

````python
from unsloth.chat_templates import get_chat_template

# tokenizer = get_chat_templateÔºà
# ‰ª§ÁâåÔºå
# chat_template =‚Äú chatml‚ÄùÔºåÔºÉÊîØÊåÅZephyrÔºåChatmlÔºåMistralÔºåLlamaÔºåÁæäÈ©ºÔºåVicunaÔºåVicuna_oldÔºåUnsploth
# Êò†Â∞Ñ= {‚ÄúËßíËâ≤‚ÄùÔºö‚ÄúÊù•Ëá™‚ÄùÔºå‚Äú content‚ÄùÔºö‚Äú value‚ÄùÔºå‚Äú user‚ÄùÔºö‚Äú human‚ÄùÔºå‚ÄúÂä©Êâã‚ÄùÔºö‚Äú gpt‚Äù}ÔºåÔºÉsharegptÊ†∑Âºè
# map_eos_token = trueÔºåÔºÉmaps <| im_end |> to </s>ËÄå‰∏çÊòØ
# Ôºâ

def formatting_prompts_func(examples):
    convos = examples["messages"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
    return { "text" : texts, }
pass

from datasets import load_dataset, Dataset
dataset = Dataset.from_json("/content/data/final/ai_meta_com_qa_pairs_cleaned_ft.json")
dataset = dataset.map(formatting_prompts_func, batched = True,)
````

XX_Markdown_link_xx‰∏äÁöÑËÅäÂ§©Ê®°Êùø‰∏äÁöÑÊõ¥Â§ö‰ø°ÊÅØ

````python
unsloth_template = \
    "{{ bos_token }}"\
    "{{ 'You are a helpful assistant to the user\n' }}"\
    "{% for message in messages %}"\
        "{% if message['role'] == 'user' %}"\
            "{{ '>>> User: ' + message['content'] + '\n' }}"\
        "{% elif message['role'] == 'assistant' %}"\
            "{{ '>>> Assistant: ' + message['content'] + eos_token + '\n' }}"\
        "{% endif %}"\
    "{% endfor %}"\
    "{% if add_generation_prompt %}"\
        "{{ '>>> Assistant: ' }}"\
    "{% endif %}"
unsloth_eos_token = "eos_token"

if False:
    tokenizer = get_chat_template(
        tokenizer,
        chat_template = (unsloth_template, unsloth_eos_token,), # ÊÇ®ÂøÖÈ°ªÊèê‰æõÊ®°ÊùøÂíåEOS‰ª§Áâå
        mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPTÈ£éÊ†º
        map_eos_token = True, # Âú∞Âõæ<| im_end |> to </s>
    )
````

xx_markDown_link_xx <a name =‚Äú train‚Äù> </a>
###ËÆ≠ÁªÉÊ®°Âûã
Áé∞Âú®ÔºåËÆ©Êàë‰ª¨‰ΩøÁî®huggingface trlÁöÑ‚Äú sfttrainer‚ÄùÔºÅËøôÈáåÊõ¥Â§öÊñáÊ°£Ôºö[link text](https://)„ÄÇÊàë‰ª¨ÊâßË°å60‰∏™Ê≠•È™§Êù•Âä†Âø´ÈÄüÂ∫¶Ôºå‰ΩÜÊòØÊÇ®ÂèØ‰ª•Â∞Ü`num_train_epochs = 1`ËÆæÁΩÆ‰∏∫ÂÆåÊï¥ËøêË°åÔºåÁÑ∂ÂêéÂÖ≥Èó≠`max_steps = none‚Äú„ÄÇÊàë‰ª¨ËøòÊîØÊåÅTRLÁöÑ‚Äú Dpotrainer‚ÄùÔºÅ

````python
from trl import SFTConfig, SFTTrainer
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    packing = False, # ÂèØ‰ª•‰ΩøÁü≠Â∫èÂàóÊõ¥Âø´Âú∞ËÆ≠ÁªÉ5ÂÄç„ÄÇ
    args = SFTConfig(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Â∞ÜÂÖ∂Áî®‰∫éWandbÁ≠â
    ),
)
````

````python
# @TitleÊòæÁ§∫ÂΩìÂâçÂÜÖÂ≠òÁªüËÆ°
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
````

````python
trainer_stats = trainer.train()
````

````python
# @TitleÊòæÁ§∫ÊúÄÁªàÂÜÖÂ≠òÂíåÊó∂Èó¥ÁªüËÆ°
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory / max_memory * 100, 3)
lora_percentage = round(used_memory_for_lora / max_memory * 100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(
    f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training."
)
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
````
