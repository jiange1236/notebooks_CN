### 推理努力
OpenAI的“ GPT-oss”模型包含一项功能，该功能允许用户调整模型的“推理工作”。这使您可以控制模型的性能与其响应速度（延迟）之间的权衡，该速度（延迟）通过该模型将使用的令牌数量来思考。

-----

“ GPT-oss”模型提供了三个不同级别的推理工作，您可以选择：

***低**：针对需要非常快速响应并且不需要复杂的多步推理的任务进行了优化。
***中**：性能与速度之间的平衡。
***高**：为需要它的任务提供最强的推理性能，尽管这会导致更高的延迟。

```python
from transformers import TextStreamer

messages = [
    {"role": "user", "content": "Solve x^5 + 3x^4 - 10 = 3."},
]
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True,
    return_tensors = "pt",
    return_dict = True,
    reasoning_effort = "low", # **新！**将推理努力设置为低，中或高
).to(model.device)

_ = model.generate(**inputs, max_new_tokens = 512, streamer = TextStreamer(tokenizer))
```

将``clunching_feft''更改为``介质''将使模型思考更长的时间。我们必须增加“ max_new_tokens”才能占据生成的令牌的数量，但它将给出更好，更正确的答案

```python
from transformers import TextStreamer

messages = [
    {"role": "user", "content": "Solve x^5 + 3x^4 - 10 = 3."},
]
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True,
    return_tensors = "pt",
    return_dict = True,
    reasoning_effort = "medium", # **新！**将推理努力设置为低，中或高
).to(model.device)

_ = model.generate(**inputs, max_new_tokens = 1024, streamer = TextStreamer(tokenizer))
```
